```python
byty = sreality[sreality["Kategorie"] == "Byty"]
byty
```

```python
byty.groupby(['Okres', 'Typ']).sum()
```

```python
byty = byty.drop(columns=['Unnamed: 0'])
byty
```

```python
byty_filtr = byty.groupby(['Okres', 'Typ']).sum().unstack()
```

```python
byty_filtr
```

```python
byty_filtr['prodej: cena/metr2'] = byty_filtr['Celková cena', 'Prodej']/byty_filtr['Užitná plocha', 'Prodej']
byty_filtr['pronájem: cena/metr2'] = byty_filtr['Celková cena', 'Pronájem']/byty_filtr['Užitná plocha', 'Pronájem']
byty_filtr['návratnost'] = byty_filtr['prodej: cena/metr2']/ (12*byty_filtr['pronájem: cena/metr2'])
```

```python
byty_filtr.sort_values('návratnost')
```

```python
byty_filtr_mean = byty.groupby(['Okres', 'Typ']).mean().unstack()
byty_filtr_mean
```

```python
byty_filtr_mean['prodej: cena/metr2'] = byty_filtr_mean['Celková cena', 'Prodej']/byty_filtr_mean['Užitná plocha', 'Prodej']
byty_filtr_mean['pronájem: cena/metr2'] = byty_filtr_mean['Celková cena', 'Pronájem']/byty_filtr_mean['Užitná plocha', 'Pronájem']
byty_filtr_mean['návratnost'] = byty_filtr_mean['prodej: cena/metr2']/ (12*byty_filtr_mean['pronájem: cena/metr2'])
```
